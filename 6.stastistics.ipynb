```python
# Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example


# A confidence interval (CI) is a statistical range or interval that is used 
# to estimate a population parameter (such as a mean, proportion, or standard
# deviation) from a sample of data. It provides a range of values within 
# which we are reasonably confident that the true population parameter
# lies, based on the sample data and a specified level of confidence.

# A confidence interval is typically expressed as:.

# Estimate  ± Margin of Error

# Example: Confidence Interval for Mean
# Suppose you are interested in estimating the average height of a population
# of adults. You collect a random sample of 100 individuals from the 
# population and find that the sample mean height is 170 cm. You want
# to calculate a 95% confidence interval for the true population mean height.


# Steps to calculate the confidence interval:

# Determine the level of confidence (α): In this case, it's 95%, so α=0.05.

# Find the critical value(s) from the standard normal distribution (Z-distribution) 
# corresponding to the chosen level of confidence. For a 95% confidence
# interval, the critical values are approximately -1.96 and 1.96. 
# You can obtain these values from a Z-table or calculator.

# Calcurlate the standard erro of the sample mean (SE) using the sample 
# standard deviation (s) and the sample size (n):
    
# SE= s/root of n

# Calculate the margin of error (ME) by multiplying the standard error by 
# the critical value:

# ME = Zα/2 (SE)

# Zα/2 is 1.96 for a 95% confidence interval

# Finally, construct the confidence interval by adding and subtracting 
# the margin of error from the sample mean:
```
